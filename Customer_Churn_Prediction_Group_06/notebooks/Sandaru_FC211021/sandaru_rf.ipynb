```python
import numpy as np
import pandas as pd #To process the dataset
```

```python
#Load the dataset from the repository
df = pd.read_csv("https://raw.githubusercontent.com/Ganindu-Deshapriya/Customer_Churn_Prediction_Group_06/refs/heads/main/Customer_Churn_Prediction_Group_06/dataset/churn_data.csv")
df.head()
```

```python
# Check for duplicate rows and remove, keeping the first occurrence
print(df.duplicated().sum())
```

```python
# Check for the missing values
missing_values_count = df.isnull().sum()
print(missing_values_count)
```

```python
# create a copy of the dataset
df_cleaned = df.copy()
```

```python
numerical_cols = df.select_dtypes(include=['float64']).columns
print(numerical_cols)
categorical_cols = df.select_dtypes(include=['object']).columns
print(categorical_cols)
```

```python
# Map binary variables
df_cleaned['Partner'] = df_cleaned['Partner'].map({'Yes': 1, 'No': 0})
df_cleaned['Dependents'] = df_cleaned['Dependents'].map({'Yes': 1, 'No': 0})
df_cleaned['PhoneService'] = df_cleaned['PhoneService'].map({'Yes': 1, 'No': 0})
df_cleaned['OnlineSecurity'] = df_cleaned['OnlineSecurity'].map({'Yes': 1, 'No': 0})
df_cleaned['OnlineBackup'] = df_cleaned['OnlineBackup'].map({'Yes': 1, 'No': 0})
#---
df_cleaned['DeviceProtection'] = df_cleaned['DeviceProtection'].map({'Yes': 1, 'No': 0})
df_cleaned['TechSupport'] = df_cleaned['TechSupport'].map({'Yes': 1, 'No': 0})
df_cleaned['StreamingTV'] = df_cleaned['StreamingTV'].map({'Yes': 1, 'No': 0})
df_cleaned['StreamingMovies'] = df_cleaned['StreamingMovies'].map({'Yes': 1, 'No': 0})
df_cleaned['PaperlessBilling'] = df_cleaned['PaperlessBilling'].map({'Yes': 1, 'No': 0})
df_cleaned['Churn'] = df_cleaned['Churn'].map({'Yes': 1, 'No': 0})

df_cleaned.head()
```
